```python
import pandas as pd
import numpy as np
```

```python
csv_file = "Resources/covid_cases.csv"
csv_file_df = pd.read_csv(csv_file, encoding="ISO-8859-1")
csv_file_df
```

```python
date_df = csv_file_df.rename(columns={"ï»¿Date" : "date"})
```

```python
date_df.dtypes
```

```python
date_df['date']=pd.to_datetime(date_df['date'])
```

```python
date_df
```

```python
mask = (date_df['date'] > '2021-12-01') & (date_df['date'] <= '2022-01-03')
```

```python
omicron_df = date_df.loc[mask]
```

```python
omicron_df
```

```python
omicron_df = omicron_df[['date', 'Cases - Total', 'Hospitalizations - Total']]
```

```python
omicron_df=omicron_df.replace(',','', regex=True)
```

```python
omicron_df['Cases - Total']=pd.to_numeric(omicron_df['Cases - Total'])

omicron_df.dtypes
```

```python
omicron_df['Percent_Hospitalized'] = (omicron_df['Hospitalizations - Total']/omicron_df['Cases - Total'])*100
```
